```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

# creating initial dataframe
dados = pd.read_csv('./dados.csv')
dados.describe()
```

```python
# creating instance of labelencoder
labelencoder = LabelEncoder()
# Assigning numerical values and storing in another column
dados['regiao_cat'] = labelencoder.fit_transform(dados['regiao'])
dados['cargo_cat'] = labelencoder.fit_transform(dados['cargo'])
```

```python

# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')
# passing regiao_cat column (label encoded values of regiao)
regiao_df = pd.DataFrame(enc.fit_transform(dados[['regiao_cat']]).toarray(), columns=['regiao_nordeste', 'regiao_norte', 'regiao_sudeste', 'regiao_sul'])
# passing cargo_cat column (label encoded values of cargo)
cargo_df = pd.DataFrame(enc.fit_transform(dados[['cargo_cat']]).toarray(), columns=['cargo_dev', 'cargo_eng_manager', 'cargo_product_manager', 'cargo_project_manager', 'cargo_scrum_master', 'cargo_team_leader'])
# merge with main df bridge_df on key values
novo_df = dados.join(regiao_df)
novo_df = novo_df.join(cargo_df)
# TODO
## fazer tratamento para dominio
novo_df = novo_df.drop(['regiao_cat', 'cargo_cat', 'regiao', 'cargo', 'dominio'],axis=1)
```

```python
# copy the data
df_min_max_scaled = novo_df.copy()
  
# apply normalization techniques by columns
columnColaboradores = 'colaboradores'
df_min_max_scaled[columnColaboradores] = (df_min_max_scaled[columnColaboradores] - df_min_max_scaled[columnColaboradores].min()) / (df_min_max_scaled[columnColaboradores].max() - df_min_max_scaled[columnColaboradores].min())    

columnXp = 'xp'
df_min_max_scaled[columnXp] = (df_min_max_scaled[columnXp] - df_min_max_scaled[columnXp].min()) / (df_min_max_scaled[columnXp].max() - df_min_max_scaled[columnXp].min())    
  
# view normalized data
dataTypeSeries = df_min_max_scaled.dtypes
dataTypeSeries
```
